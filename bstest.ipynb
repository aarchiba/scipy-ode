```python
import numpy as np
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 144
import matplotlib.pyplot as plt

import ODE
```

```python
def rhs(t, x):
    return np.array([x[1],-x[0]])
y0 = np.array([1,0])
t0 = 0
```

```python
B = ODE.BSSolver(t0,y0,rhs, first_step=np.pi)
```

```python
sincos = ODE.Solution(t0, y0, rhs, solver_type=ODE.BSSolver, first_step=1)
```

```python
ts = np.linspace(-2*np.pi,2*np.pi,1000)
x0 = []
x0d = []
x1 = []
x1d = []
for t in ts:
    a, b = sincos(t)
    x0.append(a)
    x1.append(b)
    a, b = sincos(t,1)
    x0d.append(a)
    x1d.append(b)
    
plt.plot(ts, x0, label="x")
plt.plot(ts, x1, label="y")
plt.plot(ts, x0d, label="x'")
plt.plot(ts, x1d, label="y'")
plt.xlim(ts[0],ts[-1])
plt.ylim(-2,2)
plt.legend(loc="best")
plt.axvline(1, linestyle=":")
plt.axvline(-1, linestyle=":")
```

```python
def errs(H,order):
    B = ODE.BSSolver(t0,y0,rhs, first_step=H, max_order=order)
    B.advance()
    ts = np.linspace(B.t_last, B.t, 100)
    es = [B.value(t)[0]-np.cos(t) for t in ts]
    return np.abs(B.y[0]-np.cos(B.t)), np.amax(np.abs(es))

order=8
pwr = 2*order+2
Hs = 10**np.linspace(0,1,10)
es = np.array([errs(H, order) for H in Hs])
scale = es[len(Hs)//2,0]/Hs[len(Hs)//2]**pwr
plt.loglog(Hs, es[:,0], label="solution")
plt.loglog(Hs, es[:,1], label="interpolant")
plt.loglog(Hs, scale*Hs**(pwr), label="order %d" % pwr)
plt.xlabel("Step size")
plt.ylabel("Error")
plt.legend(loc='best')
```
